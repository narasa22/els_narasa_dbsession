```python
df = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/journal_dirty_dataset_1000.csv")
display(df)
```

```python
from pyspark.sql.functions import to_date, col

df = df.withColumn("Publication_Date",to_date(col("publication_date"),"dd-MM-yyyy"))
```

```python
from pyspark.sql.functions import try_to_timestamp, col, split

df = df.withColumn("Publication_Date", try_to_timestamp(col("Publication_Date"))).withColumn("Keywords_Array",
        split(col("Keywords"), ",\\s*"))
```
